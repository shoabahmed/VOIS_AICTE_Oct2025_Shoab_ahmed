```python
import pandas as pd

# Load the data from the Excel file
df = pd.read_excel('/content/airbnb.xlsx')

# Display the first few rows to confirm loading
display(df.head())
```

Let's create a heatmap to visualize the average price based on neighbourhood group and room type.

```python
import altair as alt

# Calculate the average price by neighbourhood group and room type
avg_price_heatmap_data = df.groupby(['neighbourhood group', 'room type'])['price'].mean().reset_index()

# Create the heatmap
heatmap = alt.Chart(avg_price_heatmap_data).mark_rect().encode(
    x=alt.X('room type:N', title='Room Type'),
    y=alt.Y('neighbourhood group:N', title='Neighbourhood Group'),
    color=alt.Color('price:Q', title='Average Price'),
    tooltip=['neighbourhood group', 'room type', alt.Tooltip('price', format='.2f')]
).properties(
    title='Average Price by Neighbourhood Group and Room Type Heatmap'
)

# Display the heatmap
heatmap.display()
```

Let's create a bubble chart to visualize the most popular properties and locations based on review metrics.

```python
import altair as alt

# Calculate the average review metrics by neighbourhood
popular_properties_data = df.groupby(['neighbourhood', 'neighbourhood group']).agg(
    avg_review_rate_number=('review rate number', 'mean'),
    total_number_of_reviews=('number of reviews', 'sum'),
    avg_reviews_per_month=('reviews per month', 'mean')
).reset_index()

# Create the bubble chart
bubble_chart = alt.Chart(popular_properties_data).mark_circle().encode(
    x=alt.X('avg_review_rate_number:Q', title='Average Review Rate'),
    y=alt.Y('total_number_of_reviews:Q', title='Total Number of Reviews'),
    size=alt.Size('avg_reviews_per_month:Q', title='Average Reviews per Month'),
    color=alt.Color('neighbourhood group:N', title='Neighbourhood Group'),
    tooltip=['neighbourhood', 'neighbourhood group', 'avg_review_rate_number', 'total_number_of_reviews', 'avg_reviews_per_month']
).properties(
    title='Popularity of Neighbourhoods based on Review Metrics'
).interactive() # Make the chart interactive for zooming and panning

# Display the bubble chart
bubble_chart.display()
```

Let's create a box plot to evaluate host performance based on their review rate and the number of listings they have.

```python
import altair as alt
import numpy as np
import pandas as pd # Import pandas

# Enable VegaFusion
alt.data_transformers.enable("vegafusion")


# Create bins for calculated_host_listings_count
bins = [0, 1, 5, 10, np.inf]
labels = ['1 listing', '2-5 listings', '6-10 listings', '11+ listings']
df['listing_count_bin'] = pd.cut(df['calculated host listings count'], bins=bins, labels=labels, right=True)

# Create the box plot
box_plot = alt.Chart(df).mark_boxplot().encode(
    x=alt.X('listing_count_bin:N', title='Number of Listings'),
    y=alt.Y('review rate number:Q', title='Review Rate Number'),
    tooltip=['listing_count_bin', 'review rate number']
).properties(
    title='Host Performance (Review Rate) by Number of Listings'
)

# Display the box plot
box_plot.display()
```
